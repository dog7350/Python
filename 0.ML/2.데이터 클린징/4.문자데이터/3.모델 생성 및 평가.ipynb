```python
import pandas as pd

df = pd.read_csv("data/loan_train_cleaning.csv")
df.head()
```

```python
df.head().T
```

## 머신러닝 변수
* label : 종속 변수 또는 정답
* feature : 독립 변수 또는 속성

```python
df.columns
```

```python
features = [ 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area' ] # 문제 (사용자의 정보)

label = 'Loan_Status' # 정답 (대출 승인 또는 거부)

X, y = df[features], df[label]
```

```python
X.head().T
```

```python
y
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

## 모델 생성 및 학습
* 모델 생성 : 변수 = DecisionTreeClassifier()
* 학습 : 변수.fit(X_train, y_train)
    - 문제와 정답을 알고리즘에 학습
* 예측 : 변수.predict(X_test)
    - 확인용 문제를 predict에 넣으면 정답을 알려준다

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
pred = dtc.predict(X_test)
pred
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```
