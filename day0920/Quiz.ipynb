### 데이터 조회

```python
import pandas as pd
df = pd.read_csv("../0.ML/1.pandas/data/1.Quiz.csv")
df
```

```python
# index 설정
index_list = ["2011", "2012", "2013", "2014", "2015", "2016", "2017"]
df.index = index_list
df
```

### 컬럼 조회

```python
# 컬럼들만 조회
df.columns
```

```python
# 경부선 KTX 열만 조회
df["경부선KTX"]
```

```python
# 경부선KTX, 호남선KTX 컬럼 조회
df[["경부선KTX", "호남선KTX"]]
```

```python
# 인덱싱 사용(loc 사용 아님) 행기준 조회, 2012 ~ 2015년도
# 인덱스 번호를 이용
df[1:5]
```

```python
# 행기준 조회, 2012 ~ 2015년도
# 해당 인덱스의 이름으로 조회
df["2012":"2015"]
```

### loc를 이용한 조회

```python
# loc 사용, 컬럼이 "경부선 KTX"의 모든행
df.loc[:, ["경부선KTX"]]
```

```python
# 2011년 데이터만 선택, loc사용
df.loc[["2011"]]
```

```python
# 2013 ~ 2016년 행데이터 선택, loc사용
df.loc["2013":"2016"]
```

```python
# 2012~2014년까지 경부선 KTX 컬럼만 조회, loc사용
df.loc["2012":"2014", ["경부선KTX"]]
```

```python
# loc 사용 안함
df[ "2012":"2014" ][["경부선KTX"]]
```

```python
# 2015~2017년까지 경부선 KTX, 호남선KTX, 동해선KTX 컬럼 조회
df[ "2015":"2017" ][["경부선KTX", "호남선KTX", "동해선KTX"]]
```

```python
# isin 사용
# 경부선 등급이 'B', 'C'인 데이터 조회
df[ df["경부선등급"].isin(["B", "C"]) ]
```

```python
# 위 조건을 만족하는 경부선KTX만 조회
df[ df["경부선등급"].isin(["B", "C"]) ][["경부선KTX"]]
```
