## 결측 데이터 치환

```python
# 데이터 로드 & 결측치 확인
df = pd.read_csv("data/bicycle.csv")
df.isnull().sum()
```

```python
# 모든 결측치를 0으로 치환, 자기 자신은 변경 X
df.fillna(0)
```

```python
# 모든 결측치를 이전 값으로 치환
df.fillna(method='pad')
```

```python
# 이용거리 결측치 열을 이용거리 평균으로 치환, 자기 자신 변경
df["이용거리"].fillna(df["이용거리"].mean(), inplace=True)
df
```

```python
# 모든 결측치를 각 열의 평균으로 처리
df.fillna(df.mean(), inplace=True)
```

```python
# 이용거리를 최빈값으로 치환
df = pd.read_csv("data/bicycle.csv")
data = pd.DataFrame( df[ df["이용거리"].notnull() ]["이용거리"] )
mod = data.value_counts().max()

df["이용거리"].fillna(mod, inplace=True)
df
```

```python
# 결측 개수 확인, 이용거리 결측값이 없어야 한다
df.isnull().sum()
```
